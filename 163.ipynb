```python
#167
import csv
f=open("subwayfee.csv")
data=csv.reader(f)
header=next(data)
max=0
rate=0
for row in data:
    for i in range(4,8):
        row[i]=int(row[i])
    if row[6]!=0:
        rate=row[4]/row[6] #row[6]값이 0인 경우 존재->에러발생
        if rate>max:
            max=rate
            print(row,round(rate,2))#업데이트 될 때마다 출력
print(max)
        
        
```

```python
import csv
f=open("subwayfee.csv")
data=csv.reader(f)
header=next(data)
max=0
rate=0
for row in data:
    for i in range(4,8):
        row[i]=int(row[i])
    if row[6]!=0 and (row[4]+row[6])>100000:
        rate=row[4]/(row[4]+row[6]) #유무임 승차인원을 합해서 100,000명 이상인 경우에 계산
        if rate>max:
            max=rate
            print(row,round(rate,2))#업데이트 될 때마다 출력
print(max)
```

```python
#171
import csv
f=open("subwayfee.csv")
data=csv.reader(f)
header=next(data)
max=0
rate=0
mx_station=''

for row in data:
    for i in range(4,8):
        row[i]=int(row[i])
    if row[6]!=0 and (row[4]+row[6])>100000:
        rate=row[4]/(row[4]+row[6]) #유무임 승차인원을 합해서 100,000명 이상인 경우에 계산
        if rate>max:
            max=rate
            mx_station=row[3]+' '+row[1] #역으름row[3]+호선row[1]
            
print(mx_station,round(max*100,2))#가장 큰 비율 출력
```
