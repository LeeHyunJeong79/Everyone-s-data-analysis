```python
# age1_2 인구 공공데이터
import csv
f=open('age1.csv')
data=csv.reader(f)

header=next(data)
print("헤더:",header)

max=0   #영유아 수 
area="" #지역명

#영유아 row[3]에 저장되어 있음
for row in data:
    #print(row) #데이터 출력
    if row[3]=='':
        row[3]=0
  
    row[3]=int(row[3]) #csv파일에서 천단위구분기호를 빼서 저장후 실행
   
    if row[3] > max:
        max=row[3]
        area=row[0]

print("영유아(0세~10세)가 가장 많이 살고 있는 지역:",area)
print("영유아(0세~10세)가 가장 많이 살고 있는 지역의 영유아 수:",max)    

f.close()
```

```python
# age1_3 인구 공공데이터
import csv
f=open('age1.csv')
data=csv.reader(f)

header=next(data)
#print("헤더:",header)

max_age=0 #최고 고령화 인구수
max_rate=0 #최고 고령화 비율
area_age="" # 고령화 인구가 많은 지역명
area_rate=""# 고령화 비율이 높은 지역명
age1_sum=0
age2_sum=0
rate=0   #고령화비율(전체인구수/80세이상 인구수*100)

#row[-1]:100세 이상, row[-2]:90~99세, row[-3]:80~89 에 저장되어 있음
for row in data:
    
    if row[-1]=='':
        row[-1]=0
    if row[-2]=='':
        row[-2]=0
    if row[-3]=='':
        row[-3]=0
  
    age1_sum=int(row[-1])+int(row[-2])+int(row[-3]) #해당지역의 고령화 인구 수
    
    for a in range(3,14):
        age2_sum=age2_sum+int(row[a]) #해당지역의 전체 인구 수
        
    rate=age1_sum/age2_sum*100 #고령화 비율
    
    if rate > max_rate: #최대 고령화 비율과 지역 구하기
        max_rate=rate
        area_rate=row[0]
        
    if age1_sum > max_age: #가장 많은 고령화 인구수와 지역 구하기
        max_age=age1_sum
        area_age=row[0]

print("80세 이상 연령대가 가장 많이 살고 있는 지역:",area_age)
print("80세 이상 연령대가 가장 많이 살고 있는 지역의 80세 이상 인구수:",max_age)  
print()
print("고령화 비율이 가장 높은 지역:",area_rate)
print("고령화 비율이 가장 높은 지역의 고령화 비율:",max_rate) 

f.close()
```
