```python
import csv
f=open('gender2.csv')
data=csv.reader(f)
size=[]
name=input('찾고 싶은 지역의 이름을 알려주세요:')
for row in data:
    if name in row[0]:
        m=0
        f=0
        for i in range(101): #0~100번 반복
            m+=int(row[i+3]) #3~103번까지 데이터 가져오기
            f+=int(row[i+106]) #106~206번까지 데이터 가져오기
        break #삭제 시에도 결과 변동없음
    
size.append(m)
size.append(f)
print(size)
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
color=['crimson','darkcyan']
plt.axis('equal')
plt.pie(size,labels=['남','여'],autopct='%.1f%%',colors=color,startangle=90)
plt.title(name+' 지역의 남녀 성별 비율')
plt.show()
```
