```python
#153~156
import csv
f=open('gender.csv')
data=csv.reader(f)
m=[]
f=[]
name=input('궁금한 동네를 입력해주세요')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            
import  matplotlib.pyplot as plt
plt.scatter(m,f) #산점도 그래프 그리기
plt.show()
print("m=",m)
print("f=",f)
```

```python
import  matplotlib.pyplot as plt
plt.scatter(m,f,c=range(101),alpha=0.5,cmap='jet') #색 100가지, 투명도 0.5, 컬러맴적용
plt.colorbar() #컬러바 표시
plt.plot(range(max(m)),range(max(m)),'g') #추세선 추가
plt.show()
print("m=",m)
print("f=",f)
```

```python
import csv
import math # math함수 사용
f=open('gender.csv')
data=csv.reader(f)
m=[]
f=[]
size=[]
name=input('궁금한 동네를 입력해주세요')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append((int(row[i])+int(row[i+103])))
            
import  matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(10,5),dpi=300)
plt.title(name+'지역의 성별 인구 그래프')

plt.scatter(m,f,s=size,c=range(101),alpha=0.5,cmap='jet') #산점도 그래프 그리기
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g')
plt.xlabel('남성인구수')
plt.ylabel('여성인구수')
plt.show()
```

```python
#155~156
import csv
import math # math함수 사용
f=open('gender.csv')
data=csv.reader(f)
m=[]
f=[]
size=[]
name=input('궁금한 동네를 입력해주세요')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i])+int(row[i+103])))#sqrt사용
            
import  matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(10,5),dpi=300)
plt.title(name+'지역의 성별 인구 그래프')

plt.scatter(m,f,s=size,c=range(101),alpha=0.5,cmap='jet') #산점도 그래프 그리기
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g')
plt.xlabel('남성인구수')
plt.ylabel('여성인구수')
plt.show()
```
