```python
import csv
f=open('age.csv')
data=csv.reader(f)
result=[]
city=input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요:")
for row in data:
    if city in row[0]:
        for i in row[3:]:
            result.append(int(i))
print(result)

import matplotlib.pyplot as plt
plt.style.use('ggplot') #그래프에 스타일 적용 
plt.rc('font',family='Malgun Gothic') #한글제목을 넣기 위해 폰트를 맑은고딕으로 설정 
plt.title(city+'지역의 인구 구조') #그래프 제목
plt.plot(result)
plt.show()
```
