```python
import pandas as pd
index=pd.date_range('1/1/2021',periods=8)

import numpy as np
df=pd.DataFrame(np.random.rand(8,3),index=index,columns=list('ABC'))
df['D']=df['A']/df['B']
df['E']=np.sum(df,axis=1)
df
```

```python
df=df.sub(df['A'],axis=0)#모든 열에서 A열의 값을 뺀 값이 저장
df
```

```python
df=df.div(df['C'],axis=0) #C열의 데이터를 기준으로 열 우선 계산
df.to_csv('test.csv') #데이터 프레임을 test.csv파일로 저장
df.head()
```
