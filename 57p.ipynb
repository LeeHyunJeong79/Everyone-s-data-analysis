```python
import matplotlib.pyplot as plt
plt.title('legend')
plt.plot([10,20,30,40],label='asc')
plt.plot([40,30,20,10],label='desc')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('color') # 
#  
plt.plot([10,20,30,40],color='skyblue',label='skyblue_name')
plt.plot([40,30,20,10],color='pink',label='pink_name')
plt.legend()#
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle')
plt.plot([10,20,30,40],color='r',linestyle='--',label='dashed')
plt.plot([40,30,20,40],color='g',ls=':',label='dotted')
plt.legend()#
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker')
plt.plot([10,20,30,40,50],'r.',label='circle') #   
plt.plot([40,30,20,10,45],'g^',label='triangle up')#   
plt.legend() #
plt.show()
```
