```python
import matplotlib.pyplot as plt
plt.title('legend')
plt.plot([10,20,30,40],label='asc')
plt.plot([40,30,20,10],label='desc')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('color') #제목설정 
#그래프 그리기 
plt.plot([10,20,30,40],color='skyblue',label='skyblue_name')
plt.plot([40,30,20,10],color='pink',label='pink_name')
plt.legend()#범례표시
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle')
plt.plot([10,20,30,40],color='r',linestyle='--',label='dashed')
plt.plot([40,30,20,40],color='g',ls=':',label='dotted')
plt.legend()#범례표시
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker')
plt.plot([10,20,30,40,50],'r.',label='circle') #빨간색 원형 마커 그래프
plt.plot([40,30,20,10,45],'g^',label='triangle up')#초록색 삼각형 마커 그래프
plt.legend() #범례표시
plt.show()
```
