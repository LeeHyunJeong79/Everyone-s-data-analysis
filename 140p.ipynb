```python
#141~142
import csv
f=open('gender.csv')
data=csv.reader(f)
m=[]
f=[]
name=input('궁금한 동네를 입력해주세요.')
for row in data:
    for i in range(3,104):
        if name in row[0]:
            m.append(int(row[i])) #남성 데이터 저장하기
            f.append(int(row[i+103])) #여성 데이터 저장하기
import matplotlib.pyplot as plt
plt.plot(m,label='Male')  #남성범례표시
plt.plot(f,label='Female')#여성범례표시
plt.legend()
plt.show()
```
