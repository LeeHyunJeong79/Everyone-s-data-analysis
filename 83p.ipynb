```python
#서울의 기온 데이터를 히스토그램으로 표현한 데이터
import csv
import matplotlib.pyplot as plt
f=open('seoul2.csv')
data=csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1]!='':
        result.append(float(row[-1]))
plt.hist(result,bins=100,color='r')
plt.show()
```

```python
#히스토그램으로 8월의 데이터를 표현한 그래프
import csv
import matplotlib.pyplot as plt
f=open('seoul2.csv')
data=csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1]!='' and row[0].split('-')[1]=='08':
        result.append(float(row[-1]))
plt.hist(result,bins=100,color='r')
plt.show()
```

```python
#히스토그램으로 1월과 8월의 데이터를 표현한 그래프
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)

janury=[] #8월의 기온 데이터 저장
august=[] #1월의 기온 데이터 저장

for row in data:
    month=int(row[0].split('-')[1]) #'-' 로 구분된 값 중 2번째 값을 month에 저장
    if row[-1]!='':
        if month==8: #8월달 이라면 august=[]리스트에 저장
            august.append(float(row[-1]))
        elif month==1:#1월달 이라면 janury=[]리스트에 저장
            janury.append(float(row[-1]))
plt.figure(figsize=(15,5)) #그래프 크기 설정
plt.hist(august,bins=100, color='r', label="aug") #8월 히스토그램 그리기
plt.hist(janury,bins=100,color='b', label="jan")  #1월 히스토그램 그리기
plt.legend() #범례표시, 옵션 label과 함께
plt.show()
```
