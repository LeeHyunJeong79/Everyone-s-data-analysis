```python
#   
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
high=[]
low=[]
next(data)
for row in data:
    if row[-1]!='' and row[-2]!='':
        if(row[0].split('-')[1]=='08' and  row[0].split('-')[2]=='16'): #8 16  
            high.append(float(row[-1]))# high  
            low.append(float(row[-2]))# low  
plt.figure(figsize=(15,5))# 15,  5  
plt.title('My Birthday') #
plt.plot(high,'r',label='high')
plt.plot(low,'g',label='low')
plt.legend(loc=5)#,  
print(high)
print(low)
```

```python
#   
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
high=[]
low=[]
next(data)

for row in data:
    if row[-1]!='' and row[-2]!='':
        birthday=row[0].split('-')
        if(birthday[1]=='02' and  birthday[2]=='14'): #2 14  
            high.append(float(row[-1]))# high  
            low.append(float(row[-2]))# low  
            
plt.figure(figsize=(15,5))# 15,  5  
plt.rc('font',family='Malgun Gothic') #  
plt.rcParams['axes.unicode_minus']=False#    
plt.title('    ') #
plt.plot(high,'r',label='high')
plt.plot(low,'g',label='low')
plt.legend(loc=5)#,  
print(high)
print(low)
plt.show()
```
