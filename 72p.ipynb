```python
#내 생일의 기온변화 그래프
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
high=[]
low=[]
next(data)
for row in data:
    if row[-1]!='' and row[-2]!='':
        if(row[0].split('-')[1]=='08' and  row[0].split('-')[2]=='16'): #8월 16일 이라면 
            high.append(float(row[-1]))#최고기온을 high리스크에 저장 
            low.append(float(row[-2]))#최저기온을 low리스크에 저장 
plt.figure(figsize=(15,5))#가로 15인치, 세로 5인치로 설정 
plt.title('My Birthday') #제목표시
plt.plot(high,'r',label='high')
plt.plot(low,'g',label='low')
plt.legend(loc=5)#범례표시, 위치표시 
print(high)
print(low)
```

```python
#내 생일의 기온변화 그래프
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
high=[]
low=[]
next(data)

for row in data:
    if row[-1]!='' and row[-2]!='':
        birthday=row[0].split('-')
        if(birthday[1]=='02' and  birthday[2]=='14'): #2월 14일 이라면 
            high.append(float(row[-1]))#최고기온을 high리스크에 저장 
            low.append(float(row[-2]))#최저기온을 low리스크에 저장 
            
plt.figure(figsize=(15,5))#가로 15인치, 세로 5인치로 설정 
plt.rc('font',family='Malgun Gothic') #제목한글로 표시 
plt.rcParams['axes.unicode_minus']=False#마이너스 기호 깨짐 현상 방지
plt.title('내 생일의 기온 변화 그래프') #제목표시
plt.plot(high,'r',label='high')
plt.plot(low,'g',label='low')
plt.legend(loc=5)#범례표시, 위치표시 
print(high)
print(low)
plt.show()
```
