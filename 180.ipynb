```python
#180p 
#지하철 시간대별 데이터 시각화 하기

import csv
f=open('subwaytime.csv')
data=csv.reader(f)

next(data) #헤더가 2줄
next(data)

result=[]

for row in data:
    row[4:]=map(int,row[4:]) # 4열부터 숫자로 변환 
    result.append(row[10]) #출근 시간대 아침7시 승차데이터는 10번 열에 저장되어있음

print(len(result))
print(result)
```

```python
import matplotlib.pyplot as plt  #막대 그래프로 표현하기
plt.bar(range(len(result)),result)
plt.show()
```

```python
import matplotlib.pyplot as plt  #막대 그래프로 표현하기
result.sort() #오름차순으로 정렬
plt.bar(range(len(result)),result)
plt.show()
```

```python
#187
import csv
f=open('subwaytime.csv')
data=csv.reader(f)

next(data) #헤더가 2줄
next(data)

result=[]

for row in data:
    row[4:]=map(int,row[4:]) # 4열부터 숫자로 변환 
    result.append(sum(row[10:15:2])) #10,12,14 인덱스의 값을 합하여 막대그래프로 표현
                  #출근 시간대 아침7시,8시,9시 승차데이터는 10,12,14번 열에 저장되어있음

import matplotlib.pyplot as plt  #막대 그래프로 표현하기
result.sort() #오름차순으로 정렬
plt.bar(range(len(result)),result)
plt.show()
```

```python
#188 출근 시간대에 사람들이 가장 많이 타는 역
import csv
f=open("subwaytime.csv")
data=csv.reader(f)
next(data)
next(data)
mx=0 #최대값을 저장할 변수 초기화
mx_station='' #최대값을 갖는 역이름 
for row in data: #최대값 찾기
    row[4:]=map(int,row[4:])
    if sum(row[10:15:2])>mx:
        mx=sum(row[10:15:2])
        mx_station=row[3]+'('+row[1]+')'
print(mx_station, mx)
    
```
