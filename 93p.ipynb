```python
#매년 8월1일~31일까지의 데이터를 상자 그림으로 표현하기
import matplotlib.pyplot as plt
import csv
f=open('seoul2.csv')
data=csv.reader(f)
next(data)

day=[]

#for i in range(31):
 #   day.append([])
for row in data:
    if row[-1]!='':
        if row[0].split('-')[1]=='08':
            day.append(float(row[-1]))
plt.style.use('ggplot') #그래프 스타일 지정
plt.figure(figsize=(10,5),dpi=300) #그래프 크기 수정
plt.boxplot(day,showfliers=False) #아웃라이어 값 생략
print(day)
plt.show()
```

```python
#매년 8월1일~31일까지의 데이터를 일별(8월1일, 8월2일...일자별로)로 상자 그림으로 표현하기
import matplotlib.pyplot as plt
import csv
f=open('seoul2.csv')
data=csv.reader(f)
next(data)

day=[]

for i in range(31):
    day.append([])
for row in data:
    if row[-1]!='':
        if row[0].split('-')[1]=='08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
plt.style.use('ggplot') #그래프 스타일 지정
plt.figure(figsize=(10,5),dpi=300) #그래프 크기 수정
plt.boxplot(day,showfliers=False) #아웃라이어 값 생략
print(day)
plt.show()
```
